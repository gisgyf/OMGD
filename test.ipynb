```python
import pandas as pd
import omgd
import matplotlib.pyplot as plt
```

# Scale Detector

```python
# file with different spatial units
path_list = ['data/LST250.csv', 'data/LST500.csv', 'data/LST750.csv', 'data/LST1000.csv','data/LST1500.csv', 'data/LST2000.csv']
# dependent variable
Y = 'LST'
# explainatory variables
factors = ['Built', 'DEM', 'MNDWI', 'NDBI', 'NDVI', 'Roads']
# explainatory variables which is not continuous variables
type_factors = []
# class number interval
disc_interval = range(3, 8)
# quantile of the Q value
quantile = 0.8
# unit of dependent variables and independent variables
unit_list=['Â°C', '%', 'm', '', '', '', '']
```

## Scale detector for one variate

```python
%%time
scale_result, best_scale = omgd.scale_detector(path_list, Y, factors, disc_interval, type_factors, quantile=quantile, n_variates=1, random_state=0)
scale_result, best_scale
```

```python
omgd.scale_plot(scale_result, size_list=[250, 500, 750, 1000, 1500, 2000], unit='m')
```

## Scale detector for bi-variate

```python
%%time
scale_result, best_scale = omgd.scale_detector(path_list, Y, factors, disc_interval, type_factors, quantile=quantile, n_variates=2, random_state=0)
scale_result, best_scale
```

```python
omgd.scale_plot(scale_result, size_list=[250, 500, 750, 1000, 1500, 2000], unit='m')
```

## Scale detector for tri-variate

```python
%%time
scale_result, best_scale = omgd.scale_detector(path_list, Y, factors, disc_interval, type_factors, quantile=quantile, n_variates=3, random_state=0)
scale_result, best_scale
```

```python
plt.rcParams['figure.figsize'] = [11, 6]
omgd.scale_plot(scale_result, size_list=[250, 500, 750, 1000, 1500, 2000], unit='m')
```

# One step OMGD model

## One factor

```python
data = pd.read_csv(best_scale)
omgd_result = omgd.omgd(data, Y=Y, factors=factors, n_variates=1, disc_interval=disc_interval)
omgd_result
```

```python
plt.rcParams['figure.figsize'] = [6, 7]
omgd.omgd_plot(omgd_result, unit_list=unit_list)
```

## Bi-factor

```python
omgd_result = omgd.omgd(data, Y=Y, factors=factors, n_variates=2, disc_interval=disc_interval)
omgd_result
```

```python
plt.rcParams['figure.figsize'] = [12, 12]
omgd.omgd_plot(omgd_result, unit_list=unit_list)
```

## Tri-factor

```python
omgd_result = omgd.omgd(data, Y=Y, factors=factors, n_variates=3, disc_interval=disc_interval)
omgd_result
```

```python
plt.rcParams['figure.figsize'] = [10, 15]
omgd.omgd_plot(omgd_result, unit_list=unit_list, nrows=7, ncols=3)
```
